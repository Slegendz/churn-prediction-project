```python
import pandas as pd
```

```python
df_renewal = pd.read_csv('../../data/interim/renewal_calls_cleaned.csv')
df_email = pd.read_csv('../../data/interim/emails_cleaned.csv')
df_cc = pd.read_csv('../../data/interim/cc_calls_cleaned.csv')
df_billings = pd.read_csv('../../data/interim/billings_cleaned.csv')
```

```python
df_cc.info()
```

```python
df_merged = (
    df_billings
    .merge(df_renewal, on='co_ref', how='left', suffixes=('', '_renewal'))
    .merge(df_email, on='co_ref', how='left', suffixes=('', '_email'))
    .merge(df_cc, on='co_ref', how='left', suffixes=('', '_cc'))
)
```

```python
df_merged.shape
```

```python
for col in df_merged.columns:
    print(df_merged[col].value_counts(dropna=False))
    print(df_merged[col].isnull().sum())
    print("-" * 50)
```

```python
df_merged
```

```python
df_merged[df_merged['co_ref'] == 'VT6174'].describe()
```

```python

for col in df_merged.columns:
    if 'date' in col or 'month' in col:
        df_merged[col] = pd.to_datetime(df_merged[col])
        print(col)
        print(df_merged[col])
        
```

```python
df_open_outcomes = df_merged[df_merged['prospect_outcome'] == 'Open']
```

```python
df_filtered = df_merged[
    (
        ((df_merged['prospect_renewal_date'] - df_merged['call_date']).dt.days >= 0) &
        ((df_merged['prospect_renewal_date'] - df_merged['call_date']).dt.days <= 14)
    ) | (
        ((df_merged['prospect_renewal_date'] - df_merged['call_date_cc']).dt.days >= 0) &
        ((df_merged['prospect_renewal_date'] - df_merged['call_date_cc']).dt.days <= 14)
    )
    & 
    (df_merged['prospect_outcome'] != 'Open')
]
```

```python
df_filtered.shape
```

```python
for col in df_filtered.columns:
    print(df_filtered[col].value_counts(dropna=False))
    print(df_filtered[col].isnull().sum())
    print("-" * 50)
```

```python
df_filtered = df_filtered.fillna('Unknown')
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

def run_prospect_eda(df_filtered):
    """
    Automated EDA for df_filtered (78,946 rows, 130 columns)
    Target Variable: prospect_outcome
    """
    AV = AutoViz_Class()
    
    # Executing AutoViz with target-specific settings
    dft = AV.AutoViz(
        filename="",
        sep=",",
        depVar="prospect_outcome",    # Your specific target column
        dfte=df_filtered,             # Your cleaned dataframe
        header=0,
        verbose=1,                    # Provides data cleaning insights
        lowess=False,                 # Disabled for speed with 79k rows
        chart_format="html",          # Best for viewing 130 columns clearly
        max_rows_analyzed=80000,      
        max_cols_analyzed=135         # Slightly higher to ensure all 130 are included
    )
    return dft

# To run the analysis, just call:
run_prospect_eda(df_filtered)
```
