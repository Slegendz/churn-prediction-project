## Data Loading from Google Drive

This notebook handles the automated loading of raw data files from Google Drive to the local project directory.

### Process Overview:
1. Authenticate with Google Drive using PyDrive
2. Access the specified raw data folder
3. Download all files from the folder to `../data/raw/`
4. Print success messages for each downloaded file

### Requirements:
- Google Drive API credentials (client_secrets.json)
- Environment variables: CLIENT_SECRETS_PATH, DRIVE_RAW_FOLDER_ID
- .env file with the required variables

```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
import os
from dotenv import load_dotenv

load_dotenv()

gauth = GoogleAuth()
client_secrets_path = os.getenv("CLIENT_SECRETS_PATH")
gauth.LoadClientConfigFile(client_secrets_path)
gauth.LocalWebserverAuth()

drive = GoogleDrive(gauth)

folder_id = os.getenv("DRIVE_RAW_FOLDER_ID")

download_folder = '../data/raw'
os.makedirs(download_folder, exist_ok=True)

file_list = drive.ListFile({
    'q': f"'{folder_id}' in parents and trashed=false"
}).GetList()

for file in file_list:
    file_name = file['title']
    file_path = os.path.join(download_folder, file_name)
    file.GetContentFile(file_path)
    print(f"{file_name} downloaded successfully!")
```
