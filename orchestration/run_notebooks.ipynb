```python
import subprocess

notebooks = [
    r"../orchestration/data_loading.ipynb",
    r"../notebooks/01_eda/billings_eda.ipynb",
    r"../notebooks/01_eda/cc_calls_eda.ipynb",
    r"../notebooks/01_eda/emails_eda.ipynb",
    r"../notebooks/01_eda/renewal_calls_eda.ipynb",
    r"../orchestration/data_uploading.ipynb"
]

for nb in notebooks:
    print(f"Running {nb}...")
    
    result = subprocess.run([
        "jupyter", "nbconvert",
        "--to", "notebook",
        "--execute",
        "--inplace",
        nb
    ])
    
    if result.returncode != 0:
        print(f"Error in {nb}")
        break
    else:
        print(f"Finished {nb}")
```
